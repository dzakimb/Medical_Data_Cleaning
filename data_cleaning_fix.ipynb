## Get rid of the column

I think 91 column is too much for our data. I think we can get rid of the data we don't need so much like too much null data or just have one unique data or just id data. I think it's nice to have global column to drop the column that useful

```python
NOUSE_COLUMNS = []
```

### Make DataFrame for all the data

I think we can see the column using DataFrame format to see:
1. name of column
2. total_rows
3. total_null_data
4. total_unique_data
5. dtype_data

```python
import pandas as pd

def get_column_df(df):
  name = []
  total_rows = []
  total_null_data = []
  total_unique_data = []
  dtype_data = []
  length_of_df = len(df)
  for column in df.columns:
    name.append(column)
    total_rows.append(len(df))
    total_null_data.append(df[column].isnull().sum())
    total_unique_data.append(len(df[column].unique()))
    dtype_data.append(str(df[column].dtype))
  dict_columns = {
      'name':name,
      'total_rows':total_rows,
      'total_null_data':total_null_data,
      'total_unique_data':total_unique_data,
      'dtype_data':dtype_data
  }
  return pd.DataFrame.from_dict(dict_columns)
```

```python
column_info_df = get_column_df(df)
column_info_df
```

### No Null data

Let's see the column with no null data. Because it easier to see them. We can see if the data is useful enough for us

```python
no_null_df = column_info_df.loc[column_info_df['total_null_data'] == 0]
no_null_df.reset_index(drop=True)
```

#### Deleted the all-unique and one-unique data

So here we got 20 column that not have null data. As we can see, there are columns that have all-unique data and just one unique data. I think it doesn't have so much impact for our data like `id` or `program year`

```python
temp_df = no_null_df.loc[(no_null_df['total_unique_data'] == LENGTH_DF) | (no_null_df['total_unique_data'] == 1)]
temp_df
```

```python
NOUSE_COLUMNS = []
```

Let's drop that temp_df to our `NOUSE_COLUMNS`

```python
def add_column_to_nouse_columns(df_want_to_deleted : pd.DataFrame, list_column=NOUSE_COLUMNS):
  for column in list(df_want_to_deleted['name']):
    if column not in list_column:
      list_column.append(column)

  return list_column
```

```python
NOUSE_COLUMNS = add_column_to_nouse_columns(temp_df)
NOUSE_COLUMNS
```

### Data with NULL

Now let's mess up with data with NULL so we know what can we do about it

```python
null_df = column_info_df.loc[column_info_df['total_null_data'] != 0]
null_df.reset_index(drop=True)
```

Let's sort dataframe based on how much total null data is exist

```python
null_df.sort_values('total_null_data')
```

I think we can remove the column that all data is NULL with 50% percentage of null data

```python
temp_df = null_df.loc[null_df['total_null_data'] > int(0.5*LENGTH_DF)]
temp_df.reset_index(drop=True)
```

```python
NOUSE_COLUMNS = add_column_to_nouse_columns(temp_df)
NOUSE_COLUMNS
```

I think we can delete the name column because it doesn't seem very important and just make our models confuse.

```python
temp_columns = ['Covered_Recipient_First_Name', 'Covered_Recipient_Middle_Name', 'Covered_Recipient_Last_Name']
for column in temp_columns:
  NOUSE_COLUMNS.append(column)
NOUSE_COLUMNS
```

## Leftover column

We've been removing the column quite largely. So let's see what is the leftover column and how much columns are left.

```python
left_over_column = column_info_df.loc[~column_info_df['name'].isin(NOUSE_COLUMNS)]
left_over_column
```

Let's get the DataFrame with only that features

```python
df_new = df.drop(columns=NOUSE_COLUMNS)
df_new
```

## Filling NULL Data

I think it's enough, now let's see what we can do to complete the null data. We can make a pipeline to make a simpleimputer for the categorical features. It's because the numerical features don't have missing data. We can use sklearn pipeline for this kind of job

 ### Categorical Feature

 I think we can make a pipeline for filling the numerical feature. We can fill it with SimpleImputer on scikit-learn. There is a good article about how we handle missing data for categorical features from: https://medium.com/analytics-vidhya/handle-missing-values-in-categorical-features-b7c5b073dda2:
 1. Delete the entire column
 2. Replace missing values with the mode
 3. Delete rows with null values
 4. Predict value using a Classifier Algorithm


I think I just use two ways, replace with the mode and with the predict value. Let's grab the feature of the categorical feature

```python
categorical_features = list(left_over_column['name'].loc[(left_over_column['dtype_data']=='object') & (left_over_column['total_null_data'] != 0)])
categorical_features
```

### Replace NaN with Mode

This is the way we do the replace NaN data using mode data.

```python
df_replace_mode = df_new.copy()

def replace_with_mode(df):
  for feature in categorical_features:
    replace_with = str(df[feature].mode())
    df[feature].fillna(replace_with, inplace=True)

replace_with_mode(df_replace_mode)

df_replace_mode.isnull().sum()


```

The last we can delete Covered_Recipient_Profile_ID, and Covered_Recipient_Profile_NPI because it's not significant

```python
df_cleaned = df_replace_mode.drop(['Covered_Recipient_Profile_ID', 'Covered_Recipient_NPI'], axis=1)
df_cleaned
```
